```python
import wfdb
import matplotlib.pyplot as plt
import numpy as np

# Nome do ficheiro sem extensão
record_name = "100001_PPG"

# Ler o sinal
signal, fields = wfdb.rdsamp(record_name)

# Verificar dimensões do sinal
print("Dimensões do sinal antes da correção:", signal.shape)

# Achatar o array se tiver apenas uma linha
if signal.shape[0] == 1:
    signal = signal.flatten()

# Verificar novamente
print("Dimensões do sinal após a correção:", signal.shape)
print("Primeiros 10 valores do sinal:", signal[:10])
print("Últimos 10 valores do sinal:", signal[-10:])
# Plot do sinal corrigido
plt.figure(figsize=(10, 4))
plt.plot(signal, label="Sinal PPG", color="blue")
plt.xlabel("Amostras")
plt.ylabel("Amplitude")
plt.title("Sinal PPG de 100001, com qualidade")
plt.legend()
plt.grid()
plt.show()
```

```python
# Plot dos primeiros 3 segundos
plt.plot(signal[:fields["fs"] * 3], label="Sinal PPG", color="blue")
```
